```python
def remove_stopwords(text):
    words = text.split()
    for word in  text.split():
        if word.lower() in stopwords:
            words.remove(word)
    return ' '.join(words)

docs_df['processed_text'] = docs_df['raw_text'].apply(remove_stopwords)
```

```python
suffixes = ['ing', 'ed', 'es', 's', 'ly', 'er', 'est', 'able', 'tion', 'ive', 'ful', 'ness']

def stem_text(text):
    words = text.split()
    stemmed_words = []
    for word in words:
        for suffix in suffixes:
            if word.endswith(suffix):
                word = word.replace(suffix, '')
                break
        stemmed_words.append(word)
    return ' '.join(stemmed_words)

docs_df['processed_text'] = docs_df['processed_text'].apply(stem_text)
print(docs_df['processed_text'])
```
